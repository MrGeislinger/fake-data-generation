```python
# Get all the SKUs
skus = (sku for sku in df_raw['SKU'])

@dataclass
class Product:
    name: str
    productline: str = None
    productcategory: str = None
    size: str = None
    price: float = field(init=False)
    sku: str = field(init=False)
        
    def __post_init__(self):
        # Make sure it's in dollars & cents
        self.price = round(10 + np.random.random() * 30, 2)
        self.sku = next(skus)
```

### Order - Data - Total Order
- Order # Same for all in same order
- Total Order is the Sum of all price in order

```python
@dataclass
class Person:
    Customer_ID: str
    Customer_Name: str
    Address: str
    State: str
    ZipCode: str
    Email: str
    Phone: str
        
@dataclass
class Order:
    Date: str
    Order_number: str
    Person: Person
    Total_Order_Amount: float = field(init=False)
    Products: list[Product] = field(init=False)
    Quantity: list[int] = field(init=False) 
    Sales_Total: list[float] = field(init=False)
        
    def __post_init__(self):
        n_items = np.random.randint(1,6)
        prods = sample(products,n_items)
        self.Products = prods
        self.Quantity = [np.random.randint(1,4) for i in range(n_items)]
        prices = [p.price for p in prods]
        self.Sales_Total = [round(q*p, 2) for q,p in zip(self.Quantity, prices)]
        self.Total_Order_Amount = round(sum(self.Sales_Total), 2)
```

```python
# Create Persons
persons = [Person(*r[3:10]) for _,r in df_raw.iterrows()]
len(persons)
```

```python
# Create Orders
orders = []
for _,r in df_raw.iterrows():
    o = Order(
        r['Date'], 
        r['Order #'],
        Person=sample(persons,1)[0]
    )
    orders.append(o)
```
